First step is to import all needed programs:

```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd
```

Second step is to form collumns with only the most important aspects (Values like Latitude dont bother me):

```python
collumns = ['Province/State', 'Country/Region', 'Date',
            'Deaths', 'Cases', 'Recovered', 'Hospitalised', 
            'Vaccinated', 'First Dose', 'Second Dose']
```

Third step is vital, we merge the two dataframes (vacc and cases) as well as instructing to use preset collumns:

```python
Fulldf = (pd.read_csv("covid_19_uk_cases.csv")).merge(pd.read_csv("covid_19_uk_vaccinated.csv"))

Fulldf = Fulldf[collumns]
```

Fourth step and first visualisation is formed by creating a bargraph displaying the population per Province/State that solely have gotten one vaccination dosis. I executed this by forming a datatable, subtracting people with second dose from the first dose and then plotting the table.

# Visual 1: Population with solely one Vaccination per Province/State

```python
label = "Once vaccinated people"

data1 = Fulldf.groupby(['Province/State']).sum().reset_index()
data1[label] = data1['First Dose'] - data1['Second Dose']
ax = data1.plot.barh(x = 'Province/State',y = label,)
```

Fifth step and second visual is gotten through plotting a linegraph. Here you can see that the Province/State "Others" is not included due to realizing in the analysation that it messes up the data due to some reason. Other than that you can see that I included the figsize, axis lables and then grouped data points to eventually create line graph.

# Visual 2: Death Rate over time per Province/State

```python
data2 = Fulldf[Fulldf['Province/State'] != "Others"]

fig, ax = plt.subplots(figsize=(15,7))
ax.set_xlabel("Date")
ax.set_ylabel("Population Deaths")
data2.groupby(['Date','Province/State']).sum()['Deaths'].unstack().plot(ax = ax)
```

Sixth and last important visual shows us the recovered trend over time per Province/State. As seen I am using a similar code as above but change the input so that it shows the rate of recovered people over time.

# Visual 3: Trend of recovered corona victims over time per Province/State

```python
fig, ax = plt.subplots(figsize=(15,7))
ax.set_xlabel("Date")
ax.set_ylabel("Population Recovered")
data2.groupby(['Date','Province/State']).sum()['Recovered'].unstack().plot(ax = ax)
```
