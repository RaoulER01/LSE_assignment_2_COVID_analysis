```python
new_df = tweets[['text']]
```

```python
def findHashtags(t):
    
    return [word for word in t.split() if '#' in word]
```

```python
hashtags={}

for k in new_df['text']:
    hashs=findHashtags(k)
    for hash in hashs:
        if hash not in hashtags:
            hashtags[hash]=1
        else:
            hashtags[hash]+=1
```

```python
newdataframe=pd.DataFrame(hashtags.items(), columns=['words', 'count'])
newdataframe.head()
```

```python
import matplotlib.pyplot as plt
```

```python
def displayPlot(df,x,y):
    
    fig, ax = plt.subplots(figsize=(8, 8))

    # Plot horizontal bar graph
    df.sort_values(by=y).plot.barh(x=x,
                        y=y,
                        ax=ax,
                        color="purple")

    ax.set_title("Common Words Found")

    plt.show()
```

```python
displayPlot(newdataframe,'words','count')
```

```python
def filterHashtags(tag,filter):
    
    flag=False
    for f in filter:
        if f.lower() in tag.lower():
            flag=True
    return flag
```

```python
corona_related=newdataframe[newdataframe.words.apply(lambda x:filterHashtags(x,['corona','covid','virus']))]
```

```python
displayPlot(corona_related,'words','count')
```
